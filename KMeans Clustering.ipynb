## K-Means Clustering

Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing the dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[: , [3, 4]].values
X
```

Encoding the categorical feature

```python
# from sklearn.preprocessing import LabelEncoder
# encoder = LabelEncoder()
# X[: , 0] = encoder.fit_transform(X[: , 0])
# X
```

Using the elbow method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
 kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
 kmeans.fit(X)
 wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method for Optimum Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

Training the K-Means model on the dataset

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)
y_kmeans
```

Visualising the clusters

```python
plt.scatter(X[y_kmeans==0 , -2], X[y_kmeans==0 , -1], s=100, c='red', label='Cluster 1')
plt.scatter(X[y_kmeans==1 , -2], X[y_kmeans==1 , -1], s=100, c='blue', label='Cluster 2')
plt.scatter(X[y_kmeans==2 , -2], X[y_kmeans==2 , -1], s=100, c='green', label='Cluster 3')
plt.scatter(X[y_kmeans==3 , -2], X[y_kmeans==3 , -1], s=100, c='cyan', label='Cluster 4')
plt.scatter(X[y_kmeans==4 , -2], X[y_kmeans==4 , -1], s=100, c='grey', label='Cluster 5')
#plt.scatter(X[y_kmeans==5 , -2], X[y_kmeans==5 , -1], s=100, c='pink', label='Cluster 6')
plt.scatter(kmeans.cluster_centers_[: , 0], kmeans.cluster_centers_[: , 1], s=300, c='yellow', label='Centroids')
plt.title('Clusters of Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
